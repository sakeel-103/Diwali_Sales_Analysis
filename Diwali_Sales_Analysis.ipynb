# Exploratory Data Analysis

### Gender

```python
sns.countplot(x = 'Gender',data = df)
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group',y= 'Amount', hue='Age Group', data = sales_age, palette='Set2', legend=False  )
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
df.columns
```

```python
# total number of orders from top 10 states
sales_state = (df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10))

sns.set(rc={'figure.figsize': (15, 5)})
sns.barplot(
    data=sales_state,
    x='State',
    y='Orders',
    hue='State',        # set hue so palette works without warning
    palette='Set2',
    legend=False
)
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount', hue='State',        # set hue so palette works without warning
    palette='Set2',
    legend=False)
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation', hue='Occupation',   # for different colors without warning
    palette='Set2',
    legend=False)

for bars in ax.containers:
    ax.bar_label(bars)
```

### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(
    data=df,
    x='Product_Category',
    hue='Product_Category',   # Assign hue so each bar has a different color
    palette='Set2',
    legend=False
)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders', hue='Product_ID',   # for different colors without warning
    palette='Set2')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example: Group and sort data
top_categories = df.groupby('Product_Category', as_index=False)['Amount'].sum()
top_categories = top_categories.sort_values(by='Amount', ascending=False)

# Take top 8 categories only, rest as 'Others'
top_n = 8
main_categories = top_categories.head(top_n)
others_sum = top_categories['Amount'][top_n:].sum()

categories = list(main_categories['Product_Category'].values) + ["Other Categories"]
purchase_contributions = list(main_categories['Amount'].values) + [others_sum]

# Convert values to percentage
total_contribution = sum(purchase_contributions)
purchase_contributions = [val / total_contribution * 100 for val in purchase_contributions]

# Donut Chart
fig, ax = plt.subplots(figsize=(10, 8))
wedges, texts, autotexts = ax.pie(
    purchase_contributions,
    labels=categories,
    autopct='%1.1f%%',
    startangle=140,
    pctdistance=0.85,
    colors=plt.cm.Set3.colors,
    wedgeprops={'linewidth': 1, 'edgecolor': 'white'}
)

# Improve label size and color
for text in texts:
    text.set_fontsize(10)
for autotext in autotexts:
    autotext.set_fontsize(9)

# Draw white circle for donut
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# Add total contribution in the center
plt.text(0, 0, f"Top {top_n} Total:\n{sum(purchase_contributions[:-1]):.2f}%",
         fontsize=14, fontweight='bold', ha='center', va='center')

# Add title
plt.title(f"Top {top_n} Categories by Purchase Contribution (%)", fontsize=16)

plt.tight_layout()
plt.show()
```
